##Data Cleaning & Manipulation

```python
big_mart.dtypes
```

```python
big_mart['Outlet_Identifier'].value_counts()
```

```python
sns.boxplot(x = 'Item_Weight',data = big_mart,palette = "magma")
plt.show()
```

```python
big_mart['Item_Weight'].fillna(big_mart['Item_Weight'].mean(),inplace = True)
```

```python
big_mart['Outlet_Size'].fillna(big_mart['Outlet_Size'].mode()[0],inplace = True)
```

```python
big_mart.isnull().sum()
```

```python
big_mart['Item_Fat_Content'].value_counts()
```

```python
big_mart.Item_Fat_Content.replace(['LF', 'low fat'], ["Low Fat", "Low Fat" ], inplace=True)
big_mart.Item_Fat_Content.replace(['reg'], ["Regular"], inplace=True)
```
